Cell 1: Imports

```python
import torch
import matplotlib.pyplot as plt
import numpy as np
import os

# Make plots look nice and academic
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams.update({'font.size': 12})
```

Cell 2: Load the Data

```python
# Load the raw Acrobot dataset we generated earlier
dataset_path = "../data/raw/acrobot_dataset.pt"

if os.path.exists(dataset_path):
    dataset = torch.load(dataset_path)
    states = dataset['states'].numpy()
    actions = dataset['actions'].numpy()
    print(f"Loaded {states.shape[0]} timesteps of data!")
else:
    print("Dataset not found. Run scripts/01_generate_data.py first!")
```

Cell 3: Plot the Trajectories

```python
# Plot the angles (theta 1 and theta 2) over the first 200 steps
time_steps = np.arange(200) * 0.01 # Assuming dt = 0.01

plt.figure(figsize=(10, 4))
plt.plot(time_steps, states[:200, 0], label="Joint 1 Angle (Shoulder)")
plt.plot(time_steps, states[:200, 1], label="Joint 2 Angle (Elbow)")

plt.title("Acrobot Unforced Trajectory")
plt.xlabel("Time (seconds)")
plt.ylabel("Angle (radians)")
plt.legend()
plt.show()
```
