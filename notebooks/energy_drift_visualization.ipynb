Cell 2: Simulate the Rollout & Calculate Energy

```python
# For this visualization, we will simulate what the output of your evaluation script will look like.
# Physics-informed methods conserve energy, Black-box methods drift over time.

time_steps = np.linspace(0, 5, 500) # 5 seconds of rollout
true_energy = np.ones_like(time_steps) * 10.0 # True physics energy stays constant at 10 Joules

# Simulate LNN energy (very slight variance, but strictly bounded)
lnn_energy = true_energy + np.random.normal(0, 0.05, len(time_steps))

# Simulate MLP energy (drifts heavily due to cumulative integration errors)
mlp_energy = true_energy + (time_steps ** 1.5) * 0.8 + np.random.normal(0, 0.2, len(time_steps))
```

Cell 3: Plot the Energy Drift

```python
plt.figure(figsize=(10, 5))

# Plot the baselines
plt.axhline(y=10.0, color='black', linestyle='--', label='True System Energy (Conserved)')

# Plot the models
plt.plot(time_steps, lnn_energy, color='blue', alpha=0.8, label='Lagrangian NN (Physics-Informed)')
plt.plot(time_steps, mlp_energy, color='red', alpha=0.8, label='Vanilla MLP (Black-Box)')

plt.title("Energy Drift Over Long-Horizon Rollout (Acrobot)")
plt.xlabel("Rollout Time (seconds)")
plt.ylabel("Total System Energy (Joules)")
plt.legend(loc='upper left')

# This single plot clearly communicates the thesis contribution:
# The LNN respects the physics prior, the MLP does not.
plt.tight_layout()
plt.show()
```
